Задание:
1. Сгенерировать выборки объема N (N = 100, 500, 1000) из нормального
распределения с одинаковыми параметрами. Параметры задать
самостоятельно.
2. Построить квантильные графики (qqplot) и гистограммы для
построенных выборок.
3. Найти оценки числовых характеристик (выборочные среднее,
дисперсию, СКО).
4. Построить доверительные интервалы для математического ожидания
при известном значении дисперсии на уровне значимости α = 0,05.
5. Построить доверительные интервалы для математического ожидания
при неизвестном значении дисперсии (оцененном по выборке) на
уровне значимости α = 0,05.
6. Построить график зависимости точечных оценок математического
ожидания, а также левых и правых границ доверительных интервалов
от объема выборки (на одном рисунке).

```r
# 1
mu = 2 # mean
sigma = 1.5 # std
N100 = 100
N500 = 500
N1000 = 1000
data_norm_N100 = rnorm(N100, mu, sigma)
data_norm_N500 = rnorm(N500, mu, sigma)
data_norm_N1000 = rnorm(N1000, mu, sigma)
```

```r
# 2
qqnorm(data_norm_N100, col='red')
qqnorm(data_norm_N500, col='red')
qqnorm(data_norm_N1000, col='red')

hist(data_norm_N100,freq=FALSE,density=6,angle=45, col="blue")
hist(data_norm_N500,freq=FALSE,density=6,angle=45, col="blue")
hist(data_norm_N1000,freq=FALSE,density=6,angle=45, col="blue")
```

```r
# 3
cat("mean_norm_N100 = ", mean_norm_N100 <- mean(data_norm_N100), "\n")
cat("var_norm_N100 = ", var_norm_N100 <- var(data_norm_N100), "\n")
cat("std_norm_N100 = ", std_norm_N100 <- sd(data_norm_N100), "\n")
cat("\n")
cat("mean_norm_N500 = ", mean_norm_N500 <- mean(data_norm_N500), "\n")
cat("var_norm_N500 = ", var_norm_N500 <- var(data_norm_N500), "\n")
cat("std_norm_N500 = ", std_norm_N500 <- sd(data_norm_N500), "\n")
cat("\n")
cat("mean_norm_N1000 = ", mean_norm_N1000 <- mean(data_norm_N1000), "\n")
cat("var_norm_N1000 = ", var_norm_N1000 <- var(data_norm_N1000), "\n")
cat("std_norm_N1000 = ", std_norm_N1000 <- sd(data_norm_N1000), "\n")
```

```r
# 4
cat("дисперсия известна\n\n")
alpha = 0.05
p = 1 - (alpha / 2)
cat("interval_N100: (", mean_norm_N100 - ((sigma * qnorm(p)) / sqrt(N100)), ",", mean_norm_N100 + ((sigma * qnorm(p)) / sqrt(N100)), ")")
cat("\n")
cat("interval_N500: (", mean_norm_N500 - ((sigma * qnorm(p)) / sqrt(N500)), ",", mean_norm_N500 + ((sigma * qnorm(p)) / sqrt(N500)), ")")
cat("\n")
cat("interval_N1000: (", mean_norm_N1000 - ((sigma * qnorm(p)) / sqrt(N1000)), ",", mean_norm_N1000 + ((sigma * qnorm(p)) / sqrt(N1000)), ")")
cat("\n")
```

```r
# 5
cat("дисперсия неизвестна\n\n")
alpha = 0.05
p = 1 - (alpha / 2)
cat("interval_N100: (", mean_norm_N100 - ((std_norm_N100 * qnorm(p)) / sqrt(N100)), ",", mean_norm_N100 + ((std_norm_N100 * qnorm(p)) / sqrt(N100)), ")")
cat("\n")
cat("interval_N500: (", mean_norm_N500 - ((std_norm_N500 * qnorm(p)) / sqrt(N500)), ",", mean_norm_N500 + ((std_norm_N500 * qnorm(p)) / sqrt(N500)), ")")
cat("\n")
cat("interval_N1000: (", mean_norm_N1000 - ((std_norm_N1000 * qnorm(p)) / sqrt(N1000)), ",", mean_norm_N1000 + ((std_norm_N1000 * qnorm(p)) / sqrt(N1000)), ")")
cat("\n")
```

```r
# 6
sets_N <- c(N100, N500, N1000)

point_estimation_means <- c(mean_norm_N100, mean_norm_N500, mean_norm_N1000)

confidence_interval_left <- c(
  mean_norm_N100 - ((sigma * qnorm(p)) / sqrt(N100)),
  mean_norm_N500 - ((sigma * qnorm(p)) / sqrt(N500)),
  mean_norm_N1000 - ((sigma * qnorm(p)) / sqrt(N1000))
)

confidence_interval_right <- c(
  mean_norm_N100 + ((sigma * qnorm(p)) / sqrt(N100)),
  mean_norm_N500 + ((sigma * qnorm(p)) / sqrt(N500)),
  mean_norm_N1000 + ((sigma * qnorm(p)) / sqrt(N1000))
)

options(repr.plot.width=20, repr.plot.height=14)

plot(sets_N, point_estimation_means, type="b", pch=19, col="blue", ylim=c(min(confidence_interval_left) - 0.05, max(confidence_interval_right) + 0.05),
     xlab="Объем выборки", ylab="Точечная оценка мат. ожидания и границы доверительного интервала",
     main="Зависимость точечных оценок математического ожидания, а также левых и правых границ доверительных интервалов (дисперсия известна) от объема выборки")

lines(sets_N, confidence_interval_left, type="b", pch=19, col="red", lty=1)
lines(sets_N, confidence_interval_right, type="b", pch=19, col="red", lty=1)

legend("topright", legend=c("Среднее", "Левая граница", "Правая граница"),
       col=c("blue", "red", "red"), pch=19, lty=c(1, 1, 1))
```
